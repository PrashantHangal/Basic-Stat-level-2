Q2.	The current age (in years) of 400 clerical employees at an insurance claims processing center is normally distributed with mean  = 38 and Standard deviation  =6. For each statement below, please specify True/False. If false, briefly explain why.
A.	More employees at the processing center are older than 44 than between 38 and 44.


```python
mean = 38
std = 6
q2_lessthan_38 = stats.norm.cdf(38, loc = mean, scale = std)

q2_less_than_44 = stats.norm.cdf(44, loc = mean,  scale = std)

q2_betweeen_38_and_44 = (q2_less_than_44 - q2_lessthan_38)
print('The probability of employee age betweeen 38 and 44 is',np.round(q2_betweeen_38_and_44*100,2),'%') 

q2_morethan_44 = 1-stats.norm.cdf(44, loc = mean, scale = std)
print('The probability of employee age more than 44 is',np.round(q2_morethan_44*100,2),'%')

true_or_false = (q2_morethan_44 > q2_betweeen_38_and_44)
print('Answer:',true_or_false)
```

B.	A training program for employees under the age of 30 at the center would be expected to attract about 36 employees.

```python
q2b = stats.norm.cdf(30, loc = mean, scale = std)
print("""A training program for employees under the age of 30 at the center would be expected to attract about"""
      ,np.round((q2b*400),0),'employees')
```

Q4) Let X ~ N(100, 202). Find two values, a and b, symmetric about the mean, such that the probability of the random variable taking a value between them is 0.99.
